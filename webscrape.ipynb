```python
import pandas as pd
import random
import time
from parsel import Selector
from time import sleep
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
```

```python
chrome_options = Options()

PROXY = ""  #Host:PORT
service = Service(executable_path='chromedriver.exe')
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument(f'--proxy-server={PROXY}')
driver = webdriver.Chrome(service=service,options=chrome_options)
```

```python
j_type = ['data+engineer','data+analyst','data+scientist','machine+learning']
job_data=[]

for t in j_type:
    for page in range(0,270,10):

        driver.get(f'https://sg.indeed.com/jobs?q={t}&start={page}')
        time.sleep(random.uniform(8.5,10.9))

        #close the dialog box 
        try:
            close = driver.find_element(By.XPATH,'//button[@class = "css-yi9ndv e8ju0x51"]')

            close.click()

        except:
                pass

        jobs = driver.find_elements(By.XPATH,'//div[@class = "css-1m4cuuf e37uo190"]')

        for job in jobs:
            job.location_once_scrolled_into_view
            job.click()
            time.sleep(random.uniform(4.6,6.9))

            job_title = driver.find_element(By.XPATH,'//h2[@class = "jobsearch-JobInfoHeader-title css-161nklr e1tiznh50"]').text.strip()
            title = job_title.split('\n')
            job_description = driver.find_element(By.XPATH,'//div[@id = "jobDescriptionText"]').text.strip()

            data = {'Job_Title':title[0],'Description':job_description}
            job_data.append(data)
    print("Saving...")
        
df = pd.DataFrame(job_data)
df
driver.quit()
```
